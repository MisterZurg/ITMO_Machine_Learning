```python
CHAPTERS_NUM = 12
```

```python
alice_df = pd.DataFrame()

for idx in range(CHAPTERS_NUM):
    alice_df.loc[idx, "chapter"] = f"CHAPTER {idx + 1}"
    alice_df.loc[idx, "raw"] = chapters[idx]

alice_df
```

```python
# textPreprocessing - is a helper function that removes stop-words, numbers / non-alphabetic characters
def textPreprocessing(text):
    text = re.sub("[^a-zA-Z ]+", " ", text)
    text = re.sub(" +", " ", text)
    
    text = text.lower()
    
    tokens = word_tokenize(text)
    
    stoplist = set(stopwords.words("english"))
    words = [token for token in tokens if token not in stoplist and token != "alice"]
    
    lemmatizer = WordNetLemmatizer()
    lemmas = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(lemmas)
```

```python
nltk.download("punkt")
nltk.download("wordnet")
for index in range(len(alice_df)):
    alice_df.loc[index, "processed"] = textPreprocessing(alice_df.loc[index, "raw"])

alice_df
```

## 3.	Find Top 10 most important (for example, in terms of TF-IDF metric) words from each chapter in the text (not "Alice"); how would you name each chapter according to the identified tokens?

### 3.1 [CountVectorizer](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html#sklearn.feature_extraction.text.CountVectorizer)

```python
# getTopTenCV() - is a heplper func that uses CountVectorizer()
def getTopTenCV(text):
    # Learn the 'vocabulary' of the training data
    vect_bow = CountVectorizer()
    vect_bow.fit(text)

    # transform training data into a 'document-term matrix'
    data = vect_bow.transform(text)
    # examine the vocabulary and document-term matrix together
    df = pd.DataFrame(
        data.toarray(), 
        columns=vect_bow.get_feature_names_out()
    )
    return df
```

```python
wordsTopCV = getTopTenCV(alice_df.preprocessed)
```

```python
# plotHistogramm() - is a heplper func that prints bar chart
def plotHistogramm(topTen, chapter):
    fig = px.bar(topTen, title=f"10 Most Important Words for Chapter {chapter + 1}")
    fig.show()
```

```python
for chapter in range(CHAPTERS_NUM):
    top = wordsTopCV.loc[chapter, :].sort_values(ascending=False).head(10)
    plotHistogramm(top, chapter)
```

### 3.2 [TfidfVectorizer](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html) Equivalent to CountVectorizer followed by TfidfTransformer.

```python
# getTopTenTV() - is a heplper func that uses TfidfVectorizer()
def getTopTenTV(text):
    # Learn the 'vocabulary' of the training data
    vect_bow = TfidfVectorizer()
    vect_bow.fit(text)

    # transform training data into a 'document-term matrix'
    data = vect_bow.transform(text)
    # examine the vocabulary and document-term matrix together
    df = pd.DataFrame(
        data.toarray(), 
        columns=vect_bow.get_feature_names_out()
    )
    return df
```

```python
wordsTopTV = getTopTenTV(alice_df.preprocessed)
```

```python
for chapter in range(CHAPTERS_NUM):
    top = wordsTopTV.loc[chapter, :].sort_values(ascending=False).head(10)
    plotHistogramm(top, chapter)
```

## 4.	Find the Top 10 most used verbs in sentences with Alice. What does Alice do most often?

```python
# splitted_text - is the text between end of the content and end of the text from the beninging
sentences = re.split("[.?!]", splitted_text)
alice_only = []
for sentence in sentences:
    if "Alice" in sentence:
        alice_only.append(sentence)

alice_only_sent = " ".join(alice_only)
```

```python
# Tokenize the text
tokens = word_tokenize(alice_only_sent)

# Get the verb and nouns
verbs = [word for (word, pos) in nltk.pos_tag(tokens) if pos.startswith("V")]
nouns = [word for (word, pos) in nltk.pos_tag(tokens) if pos.startswith("N")]

# Get the top verb and noun
top_verb = max(set(verbs), key=verbs.count)
top_noun = max(set(nouns), key=nouns.count)

# Print the top verb and noun
print("The top verb is:", top_verb)
print("The top noun is:", top_noun)
```
